# Import Libraries

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

plt.style.use("ggplot")

sns.set_theme(style="whitegrid")
```

### 5. Load Dataset

```python
df = pd.read_csv("data/US_Top_50_Universities_2026.csv")

df.head()
```

### 6. Dataset Overview

```python
df.shape

df.info()

df.describe()

df.sample(5)
```

### 7. Missing Values

```python
df.isnull().sum()
```

### Visualization

```python
sns.heatmap(df.isnull(),cbar=False)
```

### 8. Duplicate Records

```python
df.duplicated().sum()
```

### 9. Feature Engineering

```python
df["University_Age"]=2026-df["Founded_Year"]
```

### 10. Rank Tier

```python
df["Rank_Tier"]=pd.cut(
    df["National_Rank"],
    bins=[0,10,25,50],
    labels=["Elite","Top","Upper"]
)
```

### 11. Research Category

```python
df["Research_Level"]=pd.cut(
    df["Research_Impact_Score"],
    bins=[0,80,90,100],
    labels=["Medium","High","Elite"]
)
```

### 12. University Age Category

```python
df["Age_Category"]=pd.cut(
    df["University_Age"],
    bins=[0,50,100,200,500],
    labels=[
        "Modern",
        "Established",
        "Historic",
        "Ancient"
    ]
)
```

## Visualizations
### 13. Public vs Private

```python
plt.figure(figsize=(8,5))

sns.countplot(
    data=df,
    x="University_Type"
)

plt.show()
```

### 14. Tuition Distribution

```python
plt.figure(figsize=(10,5))

sns.boxplot(
    data=df,
    x="University_Type",
    y="Tuition_Fees_USD"
)
```

```python
state=df["State"].value_counts()

sns.barplot(
    x=state.values,
    y=state.index
)
```

### 17. Research vs Employment

```python
sns.scatterplot(
    data=df,
    x="Research_Impact_Score",
    y="Graduate_Employment_Rate",
    hue="University_Type"
)
```

### 18. International Students vs Rank

```python
plt.figure(figsize=(12,8))

sns.heatmap(
    df.corr(numeric_only=True),
    annot=True,
    cmap="coolwarm"
)
```

### 20. Pair Plot

```python
sns.pairplot(
    df[
        [
            "National_Rank",
            "Research_Impact_Score",
            "Graduate_Employment_Rate",
            "Acceptance_Rate"
        ]
    ]
)
```

### 21. KMeans Clustering

```python
features=df[
[
"Research_Impact_Score",
"Graduate_Employment_Rate",
"Acceptance_Rate"
]]

scaler=StandardScaler()

X=scaler.fit_transform(features)

model=KMeans(
    n_clusters=3,
    random_state=42
)

df["Cluster"]=model.fit_predict(X)
```

### 22. Cluster Visualization

```python
sns.scatterplot(
    data=df,
    x="Research_Impact_Score",
    y="Graduate_Employment_Rate",
    hue="Cluster",
    palette="Set2"
)
```

## Key Findings

• Private universities generally charge higher tuition fees.

• Public universities have larger student populations.

• Higher research impact is associated with higher graduate employment.

• Lower acceptance rates are common among top-ranked universities.

• Universities with more international students tend to have better national rankings.

• California and Massachusetts host the largest number of top universities.

• Elite research universities form a distinct cluster in K-Means analysis.

## Conclusion

This exploratory analysis highlights the characteristics of top US universities. Research quality, graduate employment, and international diversity are strongly associated with higher rankings. Public institutions tend to enroll more students, while private institutions generally have higher tuition fees. The engineered features and clustering analysis provide additional insights into institutional performance.
